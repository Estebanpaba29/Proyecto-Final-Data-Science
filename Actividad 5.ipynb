```python
#HISTOGRAMA
df_tulua[df_tulua['LINEA']=='HY LINE BROWN']
import seaborn as sns
import matplotlib.pyplot as plt
sns.histplot(df_tulua[df_tulua['LINEA']=='HY LINE BROWN']['PRECIO PROMEDIO 2000-2023'], kde=True,label='HY LINE BROWN')
sns.histplot(df_tulua[df_tulua['LINEA']=='HY LINE BROWN TIR']['PRECIO PROMEDIO 2000-2023'], kde=True, label='HY LINE BROWN TIR')
sns.histplot(df_tulua[df_tulua['LINEA']=='HY LINE W-36']['PRECIO PROMEDIO 2000-2023'], kde=True, label='HY LINE W-36')
plt.xlabel('Precio Promedio')
plt.ylabel('Frecuencia')
plt.title('Distribución linea aves (TULUA) - Precio', fontsize=12, fontweight='bold')
plt.legend()
```

```python
import pandas as pd
from sqlalchemy import create_engine, text

pd.options.display.max_rows = 100
ruta="C:\\Users\\estep\\DATA SCIENCE\\PROYECTO1"
engine=create_engine('sqlite://') 
print(engine)
df = pd.read_excel('TABLA INFO CLIENTES.xlsx').to_sql('TABLA INFO CLIENTES', engine, if_exists='replace', index=False)

def runQuery(sql):
    result = engine.connect().execute((text(sql)))
    return pd.DataFrame(result.fetchall(), columns=result.keys())
```

```python
query_ejemplo= """
SELECT * FROM "TABLA INFO CLIENTES"
"""
t=runQuery(query_ejemplo)
```

```python
query1 = """SELECT CLIENTE, CIUDAD,DEPARTAMENTO
FROM "TABLA INFO CLIENTES"
WHERE LINEA != 'HY LINE BROWN'"""
g=runQuery(query1)
```
