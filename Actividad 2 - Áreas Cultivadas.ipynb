```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('Areas_cultivadas_y_produccion_agr_cola_en_Antioquia_desde_1990 (2).csv')
df
```

```python
#Grafico frecuencia Rubros

df['Rubro'].value_counts().plot(kind='bar', figsize=(40,6),ylabel='Frecuencia',title='INDICADORES - RUBRO',xlabel='Rubro')
```

```python
#Gráfico indicadores vs Subregiones

df['Subregion'].value_counts().plot(kind='bar', figsize=(15,6),ylabel='Frecuencia',title='INDICADORES - SUBREGIONES',xlabel='Regiones')
```

```python
#Histograma Rubros

import seaborn as sns
sns.histplot(df[df['Rubro']=='Café']['Subregion'],kde=False,label='Café')
sns.histplot(df[df['Rubro']=='Caña']['Subregion'],kde=False,label='Caña')
sns.histplot(df[df['Rubro']=='Maíz Tradicional']['Subregion'],kde=False,label='Maíz T')

plt.xlabel('Subregiones')
plt.ylabel('Frecuencia')
plt.title('Distribución Productos - Subregiones', fontsize=12, fontweight='bold')

plt.legend()
plt.xticks(rotation=90)
```

```python
#Gráfico de líneas
sns.lineplot(x="Subregion", y="Volumen Producción", data=df)
plt.xticks(rotation=90)


plt.xlabel('Subregiones')
plt.ylabel('Volumen')
plt.title('Volumen Producción vs Subregiones')


plt.show()
```
