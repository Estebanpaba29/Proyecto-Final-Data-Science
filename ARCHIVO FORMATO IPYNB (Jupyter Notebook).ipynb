```python
#Agrego columna con valores constantes.
df['TAMAÑO EMPRESA']=70
```

```python
# Defino una función que generará valores aleatorios para la columna 'tamaño empresa'. 
def generar_texto_aleatorio():
    textos_aleatorios = ['Pequeña', 'Mediana', 'Grande']
    return random.choice(textos_aleatorios)
```

```python
#aplico la función a cada fila de la columna deseada, con el método apply.Y asi generar los valores aleatorios
df['TAMAÑO EMPRESA'] = df['TAMAÑO EMPRESA'].apply(lambda x: generar_texto_aleatorio())

#ya habia redefindo la variable, solo que borre la línea de código-
```
