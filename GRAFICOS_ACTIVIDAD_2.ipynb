```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
ventas=pd.read_excel('HistoClientes.xlsx')
ventas.rename(columns={'CLIIENTE': 'CLIENTE'}, inplace=True)
ventas
```

```python
clientes=pd.read_excel('TABLA INFO CLIENTES.xlsx')
clientes
```

```python

X=ventas.iloc[1:]
columns=ventas.iloc[0].values
columns[0]='Cliente'
X.columns= columns
X.head()

X.columns

T=pd.melt(X,id_vars=['Cliente'])
T['value']=T['value'].astype(int)
T
(T.query("variable != 'TOTAL '").groupby('variable').sum(numeric_only=True)/10**6).plot(ylabel='Millones en ventas', xlabel='Año', title='Ventas por año')
```

```python
(T.groupby(by='Cliente').mean(numeric_only=True).sort_values(by='value', ascending=False).head(50)/10**6).plot(ylabel='Promedio ventas',kind='bar', figsize=(18,6))
```

```python
(T.groupby(by='Cliente').mean(numeric_only=True).sort_values(by='value', ascending=False).head(50)/10**6).plot(ylabel='Promedio ventas',kind='', figsize=(18,6))
```
